# Mean Squared Error (MSE) Loss

```python
import torch
```

$$
\text{MSE}(\hat{x}_{1:N}, x_{1:N})=\frac{1}{N\times{n}}\sum_{i=1}^{N}{||x_i-\hat{x}_i||_2^2}
$$

```python
def mse(x_hat, x) : 
    # |x_hat| = (batch_size, dim)
    # |x| = (batch_size, dim)
    y = ((x - x_hat)**2).mean()
    
    return y
```

```python
x = torch.FloatTensor([[1, 1,],
                       [2, 2]])
x_hat = torch.FloatTensor([[0, 0], 
                           [0, 0]])

print(x.size(), x_hat.size())
```

```python
mse(x, x_hat)
```

## Predefined MSE in PyTorch

```python
import torch.nn.functional as F 
```

```python
F.mse_loss(x, x_hat)
```

```python
F.mse_loss(x, x_hat, reduction = 'sum')
```

```python
F.mse_loss(x, x_hat, reduction = 'none')
```
