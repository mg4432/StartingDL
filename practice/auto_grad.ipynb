# Pytorch Autograd

```python
import torch 
```

```python
x = torch.FloatTensor([[1, 2],
                       [3, 4]]).requires_grad_(True)
```

```python
x1 = x + 2 
x2 = x - 2 
x3 = x1 * x2
y = x3.sum()
print(x1)
print(x2)
print(x3)
print(y)
```

```python
y.backward() # y가 scalar가 아니면 error 발생 
```

```python
print(x.grad)
```
