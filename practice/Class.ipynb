# 클래스, 인스턴스

`class 클래스이름 :`으로 클래스를 정의한 후 그 클래스의 성질이나 행동을 정의

둘 째 줄에 def sing(self) : 를 이용해서 함수 정의
- 이렇게 클래스 안에서 정의된 함수를 메서드(method)라고 부른다.

```python
class singer : 
    def sing(self) :
        return 'singing'
```

sing 메서드는 singer라는 클래스가 하는 행동을 정의한다.

singer라는 클래스를 통해서 taeji라는 객체 생성

`인스턴스명 = 클래스()`

```python
taeji = singer()
```

taeji에게 sing이라는 메서드 적용

singer 클래스에 속한 taeji 인스턴스도 sing을 할 수 있게 된다.

```python
taeji.sing()
```

```python
taeji
```

## 변수와 메서드

ex) 아마존 캐릭터를 클래스로 표현

```python
class amazon : 
    strength = 20
    dexterity = 25
    vitality = 20
    energy = 15
    
    def attack(self) : 
        return 'jab'
```

```python
jane = amazon()
mary = amazon()
```

```python
jane.attack()
```

위에서처럼 객체는 클래스에서 정의해준 변수와 메서드를 그대로 갖게 된다.

## self

self라는 것은 그 클래스의 객체를 가리키는데 jane과 mary가 똑같은 attack 메서드를 가지기 때문에 서로 구별하기 위해서 사용

즉, 메서드를 정의할 때는 항상 self 인자를 써준다고 생각하면 된다.

self 사용 예시

```python
class amazon : 
    strength = 20
    dexterity = 25
    vitality = 20
    energy = 15
    
    def attack(self) : 
        return 'jab'
    
    def exercise(self) :    
        self.strength += 2 
        self.dexterity += 3
        self.vitality += 1
```

이 메서드는 훈련을 하면 힘, 체력, 기술이 올라가는 것을 표현 

```python
eve = amazon()
eve.exercise()
eve.strength
```

```python
eve.exercise()
eve.strength
```

## 상속

상속이란 어떤 클래스가 다른 클래스의 성질을 물려받는 것을 의미한다.

어떤 클래스를 만들 때, 처음부터 모든 것을 새로 만들 필요 없이, 핵심적인 성질을 갖고 있는 다른 클래스로부터 상속을 받아서 조금만 손을 보면 원하는 클래스를 만들어낼 수 있다.

사람을 표현한 클래스 생성

```python
class person : 
    eyes = 2 
    nose = 1
    mouth = 1
    ears = 2
    arms = 2
    legs = 2 
    
    def eat(self) : 
        print('eating')
        
    def sleep(self) :
        print('sleeping')
        
    def talk(self) : 
        print('talking')
```

학생이라는 클래스를 새로 만들고자 할 때 학생도 사람이기 때문에 사람의 성질이나 행동을 갖고 있을 것이므로 사람이라는 클래스에 학생만의 특징을 추가하면 된다.

눈, 코, 입 등의 변수를 다시 생성할 필요가 없다.

```python
class student(person) : 
    def study(self) : 
        print('studying')
```

```python
lee = person()
```

```python
lee.mouth
```

```python
lee.talk()
```

```python
kim = student()
```

```python
kim.mouth
```

```python
kim.talk()
```

```python
kim.study()
```

## 객체 속의 객체

냉장고에 음식 넣기 
- 1. 냉장고 문을 연다.
- 2. 음식을 넣는다.
- 3. 냉장고 문을 닫는다.

```python
class fridge : 
    def __init__(self) : 
        self.isOpened = False
        self.foods = []
        
    def open(self) : 
        self.isOpened = True
        print('opened the door')
        
    def put(self, thing) : 
        if self.isOpened :
            self.foods.append(thing)
            print('put some food')
        else : 
            print('the door is closed')
            
    def close(self) : 
        self.isOpened = False
        print('closed the door')
        
class food : 
    pass
```

```python
f = fridge()
apple = food()
elephant = food()
```

```python
f.open()
f.put(apple)
f.put(elephant)
```

## __init__ 메서드(초기화)

```python
class book :
    def setdata(self, title, price, author) : 
        self.title = title
        self.price = price
        self.author = author
        
    def printdata(self) : 
        print('제목 :', self.title)
        print('가격 :', self.price)
        print('저자 :', self.author)
        
    def __init__(self) : 
        print('책 객체를 새로 생성')
```

```python
b = book()
```

```python
b.setdata('누가 내 치즈를 먹었을까' ,'300원', '미키')
```

```python
b.printdata()
```

```python
class book :
    def setdata(self, title, price, author) : 
        self.title = title
        self.price = price
        self.author = author
        
    def printdata(self) : 
        print('제목 :', self.title)
        print('가격 :', self.price)
        print('저자 :', self.author)
        
    def __init__(self, title, price, author):
        self.setdata(title, price, author)
        print('책 객체를 새로 생성')
```
