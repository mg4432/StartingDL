## Arithmetic Element-wise Operations

```python
a = torch.FloatTensor([[1, 2],
                       [3, 4]])
b = torch.FloatTensor([[2, 2,],
                       [3, 3]])
```

## Inplace Operations

```python
print(a)
print(a.mul(b)) # 결과 반환
print(a) # a는 변화가 없음
print(a.mul_(b)) # a에 결과를 입력
print(a) # a의 값이 새로 부여되었음
```

## Sum. Mean(Dimension Reducing Operations)

```python
x = torch.FloatTensor([[1,2], 
                       [3,4]])
```

```python
print(x.sum())
print(x.mean())
```

```python
# 특정 dimension에 대해서만 연산 수행
print(x.sum(dim = 0))
print(x.sum(dim = -1))
```

## Broadcast in Operations

```python
x = torch.FloatTensor([[1,2]])
y = torch.FloatTensor([[4,8]])

print(x.size())
print(y.size())
```

```python
z = x + y
print(z)
print(z.size())
```

## Tensor + scalar

```python
x = torch.FloatTensor([[1, 2], 
                       [3, 4]])
y = 1 

print(x.size())
```

```python
z = x + y
print(z)
print(z.size())
```

## Tensor + Vector

```python
x = torch.FloatTensor([[1, 2],
                       [4, 8]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y 
print(z)
print(z.size())
```

```python
x = torch.FloatTensor([[[1,2]]])
y = torch.FloatTensor([3,
                       5])

print(x.size())
print(y.size())
```

```python
z = x + y 
print(z)
print(z.size())
```

## Tensor + Tensor 

```python
x = torch.FloatTensor([[1,2]])
y = torch.FloatTensor([[3],
                       [5]])
print(x.size())
print(y.size())
```

```python
z = x + y 
print(z)
print(z.size())
```
