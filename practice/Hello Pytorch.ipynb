# Working Process

## 1. 문제 정의
- 가장 중요한 부분
- 해결하고자 하는 문제를 단계별로 나누고 simplify 해야함
- 신경망이라는 함수에 대입하기 위한 $x$와 결과값 $y$가 정의되어야 한다 
 - $y = f(x)$

## 2. 데이터 수집
- 문제 정의에 따라 정해진 $x$와 $y$
- 풀고자 하는 문제의 영역에 따라 수집 방법이 다름
 - 자연어처리, 컴퓨터비전 : crawling
 - 데이터분석 : 실제 수집한 데이터
- 필요에 따라 labeling 작업을 수행
 - 자동적으로 label이 $y$로 주어질 수도 있음 
 - 하지만 대부분의 경우, label이 따로 필요함 
 - unsupervised learning을 기대하면 X

## 3. 데이터 전처리 및 분석
- 수집된 데이터를 신경망에 넣어주기 위한 형태로 가공하는 과정 
 - 입출력 값을 정제(cleaning & normalization)
- 이 과정에서 EDA가 필요
 - 데이터에 알맞는 알고리즘을 찾기 위함 
 - 자연어처리, 컴퓨터비전의 경우에는 생략되기도 함
- 컴퓨터비전의 경우, 데이터 증강(augmentation)이 수행됨
 - e.g. rotation, flipping, shifting등의 간단한 연산

## 4. 알고리즘 적용
- 데이터에 대해 가설을 세우고, 해당 가설을 위한 알고리즘(모델)을 적용

## 5. 평가
- 문제 정의에 따른 공정하고 올바른 평가 방법 필요
 - 가설을 검증하기 위한 실험 설계
- test set 구성
 - 너무 쉽거나 어렵다면 판별력이 떨어짐
 - 실제 데이터와 가장 가깝게 구성되어야 함
- 정량적 평가 & 정성적 평가

## 6. 배포
- 학습과 평가가 완료된 모델 weights 파일을 배포
- RESTful API등을 통해 wrapping 후 배포
- 데이터 분포의 변화에 따른 모델 업데이트 및 유지/보수가 필요할 수 있음
