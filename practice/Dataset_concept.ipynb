# Objective 
주어진 데이터에 대해서 결과를 내는 가상의 함수를 **모사**하는 함수를 만드는 것 
- e.g. 주어진 숫자 그림을 보고 숫자 맞추기 


# In order to do this:
- 가상의 함수(ground-truth)를 통해 데이터쌍(x,y)를 수집해 데이터셋을 만들고 함수(e.g. 신경망)가 데이터를 통해 가상의 함수를 모방하도록 한다
 - $D = {\{(x_i, y_i)\}}^N _{i=1}$
- 이 때, x와 y는 각각 n차원과 m차원의 벡터로 표현될 수 있다
 - $x \in\mathbb{R}^n, y\in\mathbb{R}^m$

# Example : Tabular Dataset
- 여러 column으로 이루어진 테이블로 구성된 데이터셋(e.g. Excel)
 - 한 row가 한 개의 sample을 의미
 - Column 개수가 벡터의 차원을 의미
