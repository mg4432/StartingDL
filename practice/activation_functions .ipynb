# Activation Functions 

```python
import torch 
import torch.nn as nn 

import matplotlib.pyplot as plt 
```

```python
x = torch.sort(torch.randn(100)*10)[0] # values, indices 중에 values만 
x
```

## Sigmoid 
$$
\sigma(x) = \dfrac{1}{1+e^{-x}}
$$

```python
act = nn.Sigmoid()
print(act(x))
```

```python
print(torch.sigmoid(x))
```

```python
plt.plot(x.numpy(), torch.sigmoid(x).numpy())
plt.show()
```

## Hyperbolic Tangent(TanH)
$$
\text{tanh}(x) = \dfrac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
$$

```python
act = nn.Tanh()
print(act(x))
```

```python
print(torch.tanh(x))
```

```python
plt.plot(x.numpy(), torch.tanh(x).numpy())
plt.show()
```
